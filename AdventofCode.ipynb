--- Part Two ---
As you scan through the corrupted memory, you notice that some of the conditional statements are also still intact. If you handle some of the uncorrupted conditional statements in the program, you might be able to get an even more accurate result.

There are two new instructions you'll need to handle:

The do() instruction enables future mul instructions.
The don't() instruction disables future mul instructions.
Only the most recent do() or don't() instruction applies. At the beginning of the program, mul instructions are enabled.

For example:

xmul(2,4)&mul[3,7]!^don't()_mul(5,5)+mul(32,64](mul(11,8)undo()?mul(8,5))
This corrupted memory is similar to the example from before, but this time the mul(5,5) and mul(11,8) instructions are disabled because there is a don't() instruction before them. The other mul instructions function normally, including the one at the end that gets re-enabled by a do() instruction.

This time, the sum of the results is 48 (2*4 + 8*5).

Handle the new instructions; what do you get if you add up all of the results of just the enabled multiplications?



```python
# Solution part 2
import re
import requests

# Your Advent of Code session cookie (replace with your own!)
# SESSION = "YOUR_SESSION_COOKIE_HERE"

# URL of the corrupted memory input
url = "https://adventofcode.com/2024/day/3/input"

# HTTP headers including your session cookie
headers = {
    "Cookie": f"session={SESSION}",
    "User-Agent": "Python script (learning purposes)"
}

# Fetch the input
response = requests.get(url, headers=headers)
response.raise_for_status()  # Raise error if request fails

memory = response.text

# Regex patterns for all valid instructions we care about
mul_pattern = r"mul\((\d{1,3}),(\d{1,3})\)"
do_pattern = r"do\(\)"
dont_pattern = r"don't\(\)"

# Combine into one regex to scan sequentially
pattern = re.compile(f"{mul_pattern}|{do_pattern}|{dont_pattern}")

enabled = True  # mul instructions are initially enabled
total = 0

# Iterate through matches in order
for match in pattern.finditer(memory):
    if match.group(1) and match.group(2):
        # mul instruction
        if enabled:
            total += int(match.group(1)) * int(match.group(2))
    elif match.group(0) == "do()":
        enabled = True
    elif match.group(0) == "don't()":
        enabled = False

print("Sum of enabled multiplications:", total)
```
